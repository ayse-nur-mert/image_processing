```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import os
import cv2
```

We are importing the necessary libraries.

# 1) Saving daytime images



We are storing the daytime images in an array.

```python
image_paths = ["/content/sample_data/Group3_image1arzu.jpg", "/content/sample_data/Group3_image1ayse.jpg", "/content/sample_data/Group3_image1muhammed.jpg"]

# We assign the images to an array
images = []
for path in image_paths:
    image = mpimg.imread(path)
    image = cv2.resize(image, (1200, 1600))
    images.append(image)

daytime_images = np.array(images)

# displaying pictures
for i in range(daytime_images.shape[0]):
    plt.imshow(daytime_images[i])
    plt.show()
```

# 2) Saving nighttime images


```python
night_image_paths = ["/content/Group3_image2arzu.jpg", "/content/Group3_image2ayse.jpg", "/content/Group3_image2muhammet.jpg"]

# We assign the images to an array
night_images = []
for path in night_image_paths:
    image = mpimg.imread(path)
    image = cv2.resize(image, (1200, 1600))
    night_images.append(image)

nighttime_images = np.array(night_images)

# Displaying images
for i in range(nighttime_images.shape[0]):
    plt.imshow(nighttime_images[i])
    plt.show()
```

# 3) Processing of daytime images

3a) Displaying images side by side.

At this stage, we are displaying the listed images side by side on a canvas.

```python
  # Creating a figure with 3 columns in 1 row side by side
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Placing each image in the subplots
for i, ax in enumerate(axes):
    ax.imshow(daytime_images[i])
    ax.axis('off')

plt.show()
```

3b) Displaying 24x24 pixel sections of the images.

At this stage, we are taking 24x24 pixel sections starting from the 650x500 pixel position of the images.

```python
x = 650
y = 500
size = 24

# Creating a figure with 3 columns in 1 row side by side
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Placing a specific section of each image into the subplots
for i, ax in enumerate(axes):
    section = daytime_images[i][y:y+size, x:x+size]
    ax.imshow(section)
    ax.axis('off')
```

3c) Displaying the images obtained under the 3D heading in different color channels.

```python
for image in images:
    # Slicing the image
    section = image[y:y+size, x:x+size]

    # Separating color channels
    r = section[:,:,0]  # Red channel
    g = section[:,:,1]  # Green channel
    b = section[:,:,2]  # Blue channel

    # Creating a figure with 3 columns side-by-side
    fig, axes = plt.subplots(1, 3, figsize=(10, 3))

    # Placing each color channel in subplots
    axes[0].imshow(r, cmap='Reds')
    axes[0].axis('off')
    axes[0].set_title('Red')

    axes[1].imshow(g, cmap='Greens')
    axes[1].axis('off')
    axes[1].set_title('Green')

    axes[2].imshow(b, cmap='Blues')
    axes[2].axis('off')
    axes[2].set_title('Blue')

    plt.show()
```

3d) Printing the pixel values of the images displayed under the 3C heading.

At this stage, we are printing the matrix values of the 24x24 sections cut from the images.

```python
for image in images:

    # Printing the pixel values of the color channels
    print("Red Channel Pixel Values:")
    print(r)
    print("Green Channel Pixel Values:")
    print(g)
    print("Blue Channel Pixel Values:")
    print(b)

    plt.show()
```

3e) Displaying images in different color maps.

In this step, we will apply different color maps to each image. Since color maps only affect single-channel (grayscale) images, we will use the single channel of the images during the process.

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, ax in enumerate(axes):
    # Creating a list for different color maps
    color_maps = ['viridis', 'plasma','cividis']

    # Selecting a different color map for each image
    cmap = color_maps[i % len(color_maps)]

    # Displaying the image with pseudo-coloring
    ax.imshow(images[i][:, :, 0], cmap=cmap)
    ax.axis('off')

plt.show()
```

# 4) Processing of nighttime images.

4) Displaying images side by side.

In this step, we are displaying the listed images side by side on a canvas.

```python
# Creating a figure with 3 columns in 1 row side by side
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Placing each image in the subplots
for i, ax in enumerate(axes):
    ax.imshow(nighttime_images[i])
    ax.axis('off')

plt.show()
```

4a) Extracting the histograms of the images and displaying them side by side.


```python
# Creating a figure with 3 columns in 1 row side by side (for histograms)
fig_hist, axes_hist = plt.subplots(1, 3, figsize=(15, 5))

# Plotting the histogram of each image and placing them in subplots
for i, ax in enumerate(axes_hist):
    ax.hist(nighttime_images[i].ravel(), bins=256, range=(0, 256), color='gray')
    ax.set_title(f"Image {i+1} Histogram")

plt.show()
```

4b)Histogram Analysis of the Images

**Image 1 Histogram**  
The pixel density is concentrated at low values (0-50), with a distribution that decreases towards the right.  

These values suggest that the image is generally dark and may have low contrast.

**Improvement Suggestion:** The contrast can be increased. CLAHE (Contrast Limited Adaptive Histogram Equalization) or simple histogram equalization (cv2.equalizeHist()) can be applied.

**Image 2 Histogram**  
The histogram is more balanced but still skewed to the left.

The pixel values are concentrated in certain ranges, which suggests that the image may be trapped in specific brightness levels.

**Improvement Suggestion:** A slight gamma correction or histogram equalization can be applied to enhance brightness and contrast.

**Image 3 Histogram**  
The pixel values are almost entirely concentrated at low values, meaning the image is extremely dark.

Likely, significant brightness and contrast enhancement is required.

**Improvement Suggestion:** Techniques such as cv2.normalize(), CLAHE, or brightness-increasing gamma correction can be used.
